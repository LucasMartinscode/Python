Carregandos as bibliotecas


```python
import pandas as pd
import numpy as np
```

Carregando o arquivo json


```python
# Carregar o arquivo CSV no DataFrame
df_youtubers = pd.read_csv('topyoutube.csv')

# Verificar se a coluna "Total Views" existe no DataFrame
if 'Total Views' in df_youtubers.columns:
    # Remover vírgulas dos milhares
    df_youtubers['Total Views'] = df_youtubers['Total Views'].replace({',': ''}, regex=True)

    # Substituir strings vazias por NaN
    df_youtubers['Total Views'] = df_youtubers['Total Views'].replace('', np.nan)

    # Converter para float
    df_youtubers['Total Views'] = df_youtubers['Total Views'].astype(float)

    # Multiplicar os valores da coluna "Total Views" por 100 milhões
    df_youtubers['Total Views'] = df_youtubers['Total Views'] * 1e8
else:
    print("A coluna 'Total Views' não foi encontrada no DataFrame.")

# Exibir o DataFrame para verificar os resultados
print(df_youtubers)

```

Substituindo em branco por Não informado

```python
df_youtubers = df_youtubers.fillna('Não informado')

df_youtubers
```

Formatar coluna “AVG” para 2 casas decimais depois da vírgula

```python
df_youtubers['Avg'] = df_youtubers['Avg'].astype(float)

    # Arredondar os valores da coluna "AVG" para duas casas decimais
df_youtubers['Avg'] = df_youtubers['Avg'].round(2)

df_youtubers
```

Mostrar os top 10 usuários.

```python
 df_youtubers_sorted = df_youtubers.sort_values(by='Avg', ascending=False)

 top_10_youtubers = df_youtubers_sorted.head(10)

 top_10_youtubers
```

```python
 # Classificar o DataFrame pela coluna "Artist" em ordem decrescente
df_youtubers_sorted_by_artist = df_youtubers.sort_values(by='Artist', ascending=False)

    # Selecionar os primeiros 100 registros
top_100_youtubers = df_youtubers_sorted_by_artist.head(100)

top_100_youtubers
```

```python
top_100_youtubers.to_csv('top_100_youtubers.csv', index=False)
print("Novo arquivo CSV salvo com sucesso.")
```
