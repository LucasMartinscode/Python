```python
import pandas as pd #Biblioteca para análise e manipulação de dados
import random 
import plotly.express as px#Bibliotecas para análises gráficas
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv('winemag-data-130k-v2.csv') #Lendo o dataset.
datasetdropna = dataset.dropna() #Eliminando valores nulos junto as linhas tabela.
```

```python
datasetdropna.head() #Imprimindo os 5 datasets não nulos.
```



*   Amostragem aleatória simples





```python
Amostra_aleatoria_simples = datasetdropna.sample(n= 100, random_state=2) # Amostra simples/aleatórias utilizando 100 linhas usando random state para ser variado o número de amostras

Amostra_aleatoria_simples.head()
```

**Criando tabelas usando a biblioteca pandas e em seguida cálculos de estátistica descritiva por província.**



*   Cálculos estatísticos como média, desvio padrão, mín e max. 




```python
media_price = Amostra_aleatoria_simples['price'].mean() #Descobrindo a média total de preços
desvio_padrao_price = Amostra_aleatoria_simples['price'].std() #Descobrindo o desvio padrão total de preços
min_price = Amostra_aleatoria_simples['price'].min() #Descobrindo o minimo total de preços
max_price = Amostra_aleatoria_simples['price'].max() #Descobrindo o maximo total de preços

media_points = Amostra_aleatoria_simples['points'].mean() #Descobrindo a média total de pontos
desvio_padrao_points = Amostra_aleatoria_simples['points'].std() #Descobrindo o desvio padrão total de pontos
min_points = Amostra_aleatoria_simples['points'].min() #Descobrindo o minimo total de pontos
max_points = Amostra_aleatoria_simples['points'].max() #Descobrindo o maximo total de pontos

Medidas_totais_table = {                                      #Criando uma tabela para medidas totais de todas regiões retiradas na amostra.
    'Medida': ['Média', 'Desvio Padrão', 'Mínimo', 'Máximo'],
    'Medidas totais de preços': [media_price, desvio_padrao_price, min_price, max_price],
    'Medidas totais de pontos': [media_points, desvio_padrao_points, min_points, max_points]}

Tabela_resultados = pd.DataFrame(Medidas_totais_table)

Tabela_resultados.head()
```



*   Criando medidas de cada província




```python
dados_california = Amostra_aleatoria_simples[Amostra_aleatoria_simples['province'] == 'California'] #Filtrando medidas apenas da California 

media_price_california = dados_california['price'].mean()
desvio_padrao_price_california = dados_california['price'].std()
min_price_california = dados_california['price'].min()
max_price_california = dados_california['price'].max()

media_points_california = dados_california['points'].mean()
desvio_padrao_points_california = dados_california['points'].std()
min_points_california = dados_california['points'].min()
max_points_california = dados_california['points'].max()


dados_oregon = Amostra_aleatoria_simples[Amostra_aleatoria_simples['province'] == 'Oregon'] #Filtrando medidas apenas de Oregon

media_price_oregon = dados_oregon['price'].mean()
desvio_padrao_price_oregon = dados_oregon['price'].std()
min_price_oregon = dados_oregon['price'].min()
max_price_oregon = dados_oregon['price'].max()

media_points_oregon = dados_oregon['points'].mean()
desvio_padrao_points_oregon = dados_oregon['points'].std()
min_points_oregon = dados_oregon['points'].min()
max_points_oregon = dados_oregon['points'].max()

dados_washington = Amostra_aleatoria_simples[Amostra_aleatoria_simples['province'] == 'Washington'] #Filtrando medidas apenas de Washington

media_price_washington = dados_washington['price'].mean()
desvio_padrao_price_washington = dados_washington['price'].std()
min_price_washington = dados_washington['price'].min()
max_price_washington = dados_washington['price'].max()

media_points_washington = dados_washington['points'].mean()
desvio_padrao_points_washington = dados_washington['points'].std()
min_points_washington = dados_washington['points'].min()
max_points_washington = dados_washington['points'].max()

Medidas_isoladas = {                                         #Criando uma tabela isolada para medidas de cada região retiradas na amostra.
    'Medida': ['Média', 'Desvio Padrão', 'Mínimo', 'Máximo'],
    'Medidas de preços da California': [media_price_california, desvio_padrao_price_california, min_price_california, max_price_california],
    'Medidas de preços de Washington': [media_price_washington, desvio_padrao_price_washington, min_price_washington, max_points_washington],
    'Medidas dos preços de Oregon': [media_price_oregon, desvio_padrao_price_oregon, min_price_oregon, max_price_oregon],
    'Medidas de pontos da California': [media_points_california, desvio_padrao_points_california, min_points_california, max_points_california],
    'Medidas de pontos de Washington': [media_points_washington, desvio_padrao_points_washington, min_points_washington, max_points_washington],
    'Medidas dos pontos de Oregon': [media_points_oregon, desvio_padrao_points_oregon, min_points_oregon, max_points_oregon]}

Tabela_resultados = pd.DataFrame(Medidas_isoladas)

Tabela_resultados.head()
```



*   Criando gráfico de barras comparando as médias de preços de vinhos entre as províncias da Califórnia, Washington e Oregon.




```python
provincias = ['California', 'Washington', 'Oregon']
medias = [media_price_california, media_price_washington, media_price_oregon]
cores_barras = ['blue', 'green', 'red']

fig, ax = plt.subplots()
sns.barplot(x=provincias, y=medias, palette=cores_barras, ax=ax)   #Criando um gráfico de barras com média de preços de vinhos por região

ax.set_title('Média dos Preços do Vinho por Província')
ax.set_xlabel('Província')
ax.set_ylabel('Média dos preços do Vinho')

plt.savefig('Gráfico de barras.png')

plt.show()
```


*   Criando um gráfico de dispersão para plotar os pontos do vinho em relação aos preços. Cada ponto representa um vinho.

```python

fig2 = sns.scatterplot(data=Amostra_aleatoria_simples, x='price', y='points', hue='province') # Criando um gráfico estatístico com cores por província


fig2.set_title('Dispersão de Preço vs. Pontos por Região')
fig2.set_xlabel('Preço')
fig2.set_ylabel('Pontos')

plt.savefig('grafico_dispersion.png')
```



*   Criando um gráfico BoxPlot para visualização de mín e máx de cada região.




```python
fig, ax = plt.subplots()
sns.boxplot(x='province', y='price', data=Amostra_aleatoria_simples, palette='Set3')

ax.set_title('Distribuição de Preços por Província')
ax.set_xlabel('Província')
ax.set_ylabel('Preço')

plt.savefig('Gráfico boxplot.png')

plt.show()
```

**Conclusão**

Com base nos valores da média de todas regiões (42.88), desvio padrão de todas regiões (30.322935), mín e máx (13.00) e (175.00) respectivamente, podemos ter algumas inferências sobre os preços dos vinhos no futuro:

A média (42.88) pode ser considerada um valor central ou estimativa do preço médio dos vinhos. Isso significa que, em média, os preços dos vinhos podem se situar em torno desse valor no futuro.

O desvio padrão (30.322935) indica a dispersão dos preços em relação à média. Um desvio padrão maior sugere uma maior variabilidade nos preços, o que significa que os preços dos vinhos podem ter flutuações mais significativas no futuro.

O valor mínimo (13.00) representa o menor preço observado nos dados históricos. Isso indica que, no passado, houve vinhos com preços tão baixos quanto 13.00.

O valor máximo (175.00) representa o maior preço observado nos dados históricos. Isso indica que, no passado, houve vinhos com preços tão altos quanto 175.00.
